```python
import pandas as pd
import requests
from bs4 import BeautifulSoup
```

```python
reviewlist = []
```

```python
def get_soup(url):
    data  = requests.get(url).text 
    soup = BeautifulSoup(data,"html.parser")
    return soup
```

```python
def get_reviews(soup):
    reviews = soup.find_all('div',{'class':'review-content'})
    try:
        for items in reviews:
            review = {
            'Review_title' : items.find('a',{'class':'link link--large link--dark'}).text.strip(),
            'Ratings' : float(items.find('img')['alt'][0]),
            'Body' : items.find('p',{'class':'review-content__text'}).text.strip()
            }
            reviewlist.append(review)
    except:
        pass
```

```python
for i in url:
    soup = get_soup(i)
    get_reviews(soup)
```
